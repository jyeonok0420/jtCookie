```python
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
image = cv2.imread("tigers.jpg")
gray_img = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
can_image = cv2.Canny(gray_img, 100, 200)

#cv2_imshow(image)
#cv2_imshow(can_image)

kernel = np.ones((5,5), np.uint8)
#img_erdoe = cv2.erode(can_image, kernel, iterations=1)
#cv2_imshow(img_erdoe)
img_dilation = cv2.dilate(can_image, kernel, iterations=1)
#cv2_imshow(img_dilation)
img_erdoe = cv2.erode(img_dilation, kernel, iterations=1)
#cv2_imshow(img_erdoe)
display = np.hstack((can_image, img_dilation, img_erdoe))
cv2_imshow(display)
```
